```python
% matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# all the cutflow results
theory = [[64400, 46300, 24900,24300],'theory']
METAK8 = [[197177,159862,90679,85351], 'MET (AK8)']
MHTAK8 = [[172840,135440,62390,57062], 'MHT (AK8)']
MHTSmAK8 = [[1103629,396754,312001,274456], 'MHT Smeared (AK8)']
METCA = [[197177,160377,89688,84363], 'MET (CA11)']
MHTCA = [[172840,133308,61716,56391], 'MHT (CA11)']
MHTSmCA = [[1103629,810051,302656,265229], 'MHT Smeared (CA11)']
CMS = [[993373,512527,276175,245508], 'CMS']

# this function will these results side by side for comparison
def compare_Cutflow(listOfResults):
    # global fontsize formatting
    font = {'size'   : 12}
    mpl.rc('font', **font)

    # some parameters
    N = 4               # number of cuts
    ind = np.arange(N)  # the x locations for the groups
    width = 0.8/len(listOfResults)       # the width of the bars

    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(111)

    colorlist = ['r','y','g','b','m','c']
    rectlist = []

    for i in range(len(listOfResults)):
        colorval = i%len(colorlist)
        rectlist.append(ax.bar(ind+width*i, listOfResults[i][0], width, 
                               color=colorlist[colorval],label=listOfResults[i][1]))


    ax.set_ylabel('Number of Events',fontsize=20)
    ax.set_xticks(ind+width)
    ax.set_xticklabels( ('Trigger and presel.', '$MET/M_T > 0.15$', '$|\Delta\eta| > 1.1$', 
                         '$|\Delta\phi| > 1.1$') )
    ax.legend()

    def autolabel(rects):
        for rect in rects:
            h = rect.get_height()
            ax.text(rect.get_x()+rect.get_width()/2., 1.02*h, '%d'%int(h),
                    ha='center', va='bottom')

    for i in range(len(rectlist)):
        autolabel(rectlist[i])

    mpl.rc('xtick', labelsize=18) 
    mpl.rc('ytick', labelsize=18)
```

## Comparing MET, MHT, MHTSmeared with Theory and CMS (AK8)

```python
listOfR = [theory, METAK8, MHTAK8, MHTSmAK8, CMS]

compare_Cutflow(listOfR)
```

## Comparing MET, MHT, MHTSmeared with Theory and CMS (CA11)

```python
listOfR = [theory, METCA, MHTCA, MHTSmCA, CMS]

compare_Cutflow(listOfR)
```

## Comparing AK8 with CA11

```python
listOfR = [METAK8, METCA, MHTAK8, MHTCA, MHTSmAK8, MHTSmCA]

compare_Cutflow(listOfR)
```

## Comparing MET, MHT, MHTSmeared w/ Theory & CMS (CA11 w/ Theory Settings but AK4)

```python
METCAwT = [[203219,165291,92436,86948], 'MET (CA11) with Theory Settings']
MHTCAwT = [[178136,137393,63607,58119], 'MHT (CA11) with Theory Settings']
MHTSmCAwT = [[1137446,834872,311929,273356], 'MHT Smeared (CA11) with Theory Settings']

listOfR = [theory, METCAwT, MHTCAwT, MHTSmCAwT, CMS]

compare_Cutflow(listOfR)
```

## Comparing CA11 and CA11 with Theory Settings

```python
listOfR = [METCA, METCAwT, MHTCA, MHTCAwT, MHTSmCA, MHTSmCAwT]

compare_Cutflow(listOfR)
```
